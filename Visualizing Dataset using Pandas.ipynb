# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\sivas\OneDrive\Documents\sample datasets\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python
df.plot(kind = 'line', subplots = True, title = 'Ice Cream Rating')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily_Rating', ylabel = 'Rating_Score')
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Rating', xlabel = 'Daily_Rating', ylabel = 'Rating_Score')
```

```python
df['Texture Rating'].plot(kind = 'bar', title = 'Ice Cream Rating', xlabel = 'Daily_Rating', ylabel = 'Rating_Score')
```

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Rating', xlabel = 'Daily_Rating', ylabel = 'Rating_Score')
```

```python
df.plot.barh(stacked=True, title = 'Ice Cream Rating', xlabel = 'Daily_Rating', ylabel = 'Rating_Score')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Flavor Rating')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'Orange')
```

```python
df.plot.hist(bins = 49, title = 'Ice Cream Rating', xlabel = 'Daily_Rating')
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (7,7))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```

```python
df.plot.pie(y = 'Texture Rating', figsize = (10,10))
```

```python
df.plot.pie(y = 'Overall Rating', figsize = (10,10))
```

```python
df.plot.pie(subplots = True, figsize = (39,50))
```
